```python
import chromadb
chroma_client = chromadb.Client()
```

```python
chroma_client.create_collection(name="news")
```

```python
collection = chroma_client.get_collection(name="news")
```

```python
collection.add(
    documents=["Apple reported its quartely earnings today.",
               "Apple has a lot of vitamins."],
    metadatas=[{"source": "my_source"}, {"source": "my_source"}],
    ids=["id1", "id2"]
)
```

```python
collection.query(
    query_texts=["New iPhone will launch in september."],
    n_results=2
)
```

```python
collection.query(
    query_texts=["My niece loves oranges."],
    n_results=2
)
```

```python
client = chromadb.PersistentClient(path="./news_vector_db")
```

```python
from chromadb.utils import embedding_functions

ef = embedding_functions.SentenceTransformerEmbeddingFunction(
  model_name="all-distilroberta-v1"
)

collection = chroma_client.create_collection(
    name="news_v2",
    embedding_function=ef
)
```

```python
collection.add(
    documents=["Apple reported its quartely earnings today.",
               "Apple has a lot of vitamins."],
    metadatas=[{"source": "my_source"}, {"source": "my_source"}],
    ids=["id1", "id2"]
)
```

```python
collection.query(
    query_texts=["New iPhone will launch in september."],
    n_results=2
)
```
